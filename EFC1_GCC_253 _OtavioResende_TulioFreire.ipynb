##  Insertion Sort


```python
def insertionSort(A):
  comparacoes = 0
  for index in range(1,len(A)):

    currentvalue = A[index]
    position = index

    while position>0 and A[position-1] > currentvalue:
      comparacoes += 1
      A[position] = A[position-1]
      position = position-1

    A[position] = currentvalue

  return (comparacoes)

```

##  Merge Sort


```python
def mergeSort(A):
  comparacoes = 0
  if len(A)>1:
    comparacoes += 1
    mid = len(A)//2
    lefthalf = A[:mid]
    righthalf = A[mid:]

    mergeSort(lefthalf)
    mergeSort(righthalf)

    i=0
    j=0
    k=0
    while i < len(lefthalf) and j < len(righthalf):
      if lefthalf[i] < righthalf[j]:
        A[k]=lefthalf[i]
        i=i+1
      else:
        A[k]=righthalf[j]
        j=j+1
      k=k+1

    while i < len(lefthalf):
      A[k]=lefthalf[i]
      i=i+1
      k=k+1

    while j < len(righthalf):
      A[k]=righthalf[j]
      j=j+1
      k=k+1

  return comparacoes

```

##  Selection Sort


```python
def selectionSort(A):
  comparacoes = 0
  for i in range(len(A)):
    posicaoMenor = i
    for j in range(i, len(A)):
      if(A[j] < A[posicaoMenor]):
        comparacoes += 1
        posicaoMenor = j
    if(posicaoMenor != i):
      temp = A[i]
      A[i] = A[posicaoMenor]
      A[posicaoMenor] = temp

  return comparacoes
```

##  Bubble Sort


```python
def bubbleSort(A):
  comparacoes = 0
  for passnum in range(len(A)-1,0,-1):
    for i in range(passnum):
      if A[i]>A[i+1]:
        comparacoes += 1
        temp = A[i]
        A[i] = A[i+1]
        A[i+1] = temp

  return comparacoes
```

## Heap Sort

```python
from heapq import heappop, heappush

def heapSort(A):
  comparacoes = 0
  heap = []
  for element in A:
    heappush(heap, element)

  ordered = []

  while heap:
    comparacoes += 1
    ordered.append(heappop(heap))


  A = ordered

  return comparacoes
```

## Quick Sort

> pivo = *A[A-comprimento]*

```python
def quickSort(A):
   comparacoes = 0
   quickSortHelper(A,0,len(A)-1, 0)

   return comparacoes

def quickSortHelper(A,first,last, comparacoes):
  
   if first<last:
    comparacoes += 1

    splitpoint = partition(A,first,last)

    quickSortHelper(A,first,splitpoint-1)
    quickSortHelper(A,splitpoint+1,last)


def partition(A,first,last):
   pivotvalue = A[first]

   leftmark = first+1
   rightmark = last

   done = False
   while not done:

    while leftmark <= rightmark and A[leftmark] <= pivotvalue:
        leftmark = leftmark + 1
```

## Graficos Insertion Sort

```python
import matplotlib.pyplot as plt

def execInsertionSort():

  # Melhor caso (Array ordenado)

  A = []

  for i in range(5, 1005, 5):
    A.append(i)

  inicio = timeit.default_timer()
  comparacoesMelhorCaso = insertionSort(A)
  fim = timeit.default_timer()
  tempoMelhorCaso = fim - inicio

  # Pior caso (Array ordenado em ordem inversa)

  A = []

  for i in range(1000, 0, -5):
    A.append(i)


  inicio = timeit.default_timer()
  comparacoesPiorCaso = insertionSort(A)
  fim = timeit.default_timer()
  tempoPiorCaso = fim - inicio

  # Gráfico tempo de execução

  x  = ["Melhor caso", "Pior Caso"]
  y1 = [0.0, tempoMelhorCaso]
  y2 = [0.0, tempoPiorCaso]
  plt.plot(x, y1, label="Tempo melhor caso")
  plt.plot(x, y2, label="Tempo pior caso")
  plt.plot()

  plt.xlabel("Casos")
  plt.ylabel("Tempos(ms)")
  plt.title("Gráfico")
  plt.legend()
  plt.show()

  # Gráfico número de comparações

  x  = ["Melhor caso", "Pior Caso"]
  y1 = [0, comparacoesMelhorCaso]
  y2 = [0, comparacoesPiorCaso]
  plt.plot(x, y1, label="Comparações melhor caso")
  plt.plot(x, y2, label="Comparações pior caso")
  plt.plot()

  plt.xlabel("Casos")
  plt.ylabel("Comparações")
  plt.title("Gráfico")
  plt.legend()
  plt.show()

execInsertionSort()
```

Insertion Sort: pior caso e caso medio = O(n^2), melhor caso = O(n)
Merge Sort: pior, melhor e caso medio = Θ(n∗logn)
Selection Sort: pior, melhor e caso medio = O(n^2)
